## 질병에관한 전체상관분석

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
sub_col = [x for x in all_mean_columns if x not in disease]

def disease_all_corr(start, end):
  df1 = health[sub_col[start:end]]
  df2 = health[disease]

  df = pd.concat([df1,df2], axis=1)
  scaler = MinMaxScaler()
  scaler.fit(df)
  df_scaled = scaler.transform(df)
  scaled_df = pd.DataFrame(df_scaled,  columns=df.columns)
  corr_df = scaled_df.corr(method='pearson')

  plt.figure(figsize=(35,15))
  sns.heatmap(corr_df, annot=True, fmt=".2f", cmap='Blues')

disease_all_corr(0,20)
disease_all_corr(20,40)
disease_all_corr(40,60)
disease_all_corr(60,80)
disease_all_corr(80,100)
disease_all_corr(100,107)
```
